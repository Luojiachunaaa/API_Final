### 关键字搜索

```python
def place_text(keywords,types,city=None,citylimit=None,children=None,page=None,extensions='base',sig=None)->dict:
    url = 'https://restapi.amap.com/v3/place/text?parameters'
    params={
        'key':key_luo,
        'keywords':'郑州西亚斯学院',
        'types':'科教文化服务',
        'city':'郑州',
        'citylimit':True,
        'children':1,
        'page':page,
        'extensions':'all',
        'output':'json'
    }
    response = requests.get(url,params=params)
    data = response.json()
    return data
```
